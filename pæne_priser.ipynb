```python
import random
import time

N = 13650  #antallet i VM
priser = [round(random.uniform(3, 100),2) for _ in range(N*100)]


priser = np.zeros(N)

start_time = time.time()

for i, pris in enumerate(priser):
    priser[i] = pæne_priser(pris, AFRUNDINGER)
    

end_time = time.time()
total_time = end_time - start_time
print(total_time)
```
